Seleciona no portal os contratos com Maiores valores

```python
# Adiciona o parâmetro de valor, bem alto, 10 milhões
```

```python
valor = 10000000.0
```

```python
# Página inicial - os lotes de páginas correm de 500 em 500
```

```python
pagina = 500
```

```python
url_padrao = "http://compras.dados.gov.br/contratos/v1/contratos.json?valor_inicial_min=" + str(valor) + "&offset=" 
url = url_padrao + str(pagina)
```

```python
url
```

```python
codigo = 200 # código para uma página encontrada
# code 200 - OK - We received the request
```

```python
while(codigo == 200): # Enquanto o código for 200...
    r = requests.get(url)
    
    print("lote: ", pagina)
    if r.status_code == 503:
        print("503 error code")
        codigo = 503 # a página não foi encontrada então vai sair do loop
    else:
        print(url)
        contratos = r.json()['_embedded']['contratos']
        contratos_mais_caros = pd.DataFrame(contratos)
        
        if pagina == 500: # se for a primeira página
            contratos_mais_caros_final =  contratos_mais_caros
        else:
            contratos_mais_caros_final = pd.concat([contratos_mais_caros_final, contratos_mais_caros])
        
    pagina = pagina + 500
    url = url_padrao + str(pagina)
```
