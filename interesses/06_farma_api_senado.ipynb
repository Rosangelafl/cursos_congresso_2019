```python
import requests
import pandas as pd
```

Fonte: http://legis.senado.leg.br/dadosabertos/docs/ui/index.html

```python
# Esta API pede para passar um header com o tipo de dado desejado

## É o item PesquisaMateriaService
```

```python
headers = {"Accept" : "application/json"}
```

```python
prefixo = 'http://legis.senado.leg.br/dadosabertos/materia/pesquisa/lista?ano='
```

```python
# Cria lista vazia para receber resultados
projetos = []

# Percorre entre os anos de 2010 e 2018
for i in range(2010, 2019):
    ano = str(i)
    print("ANO: ", ano)
    
    url = prefixo + ano
    print(url)
    
    # Este é um exemplo de teste padrão em requests - pode ser usado sempre
    try:
        r = requests.get(url, headers=headers)
    except requests.exceptions.HTTPError as errh:
        print ("Http Error:",errh)
    except requests.exceptions.ConnectionError as errc:
        print ("Error Connecting:",errc) 
    except requests.exceptions.Timeout as errt:
        print ("Timeout Error:",errt)
    except requests.exceptions.RequestException as err:
        print ("OOps: Something Else",err)

    # Transforma o resultado em JSON
    projects = r.json()
    
    
    # Percorre o JSON a partir deste local
    for item in projects["PesquisaBasicaMateria"]["Materias"]["Materia"]:
        
        # Faz testes em cada campo de informação 
        try:
            AnoMateria = str(item['IdentificacaoMateria']['AnoMateria'])
        except KeyError:
            AnoMateria = None
            
        try:
            ApelidoMateria = item['DadosBasicosMateria']['ApelidoMateria']
        except KeyError:
            ApelidoMateria = None
            
        try:
            CodigoMateria = str(item['IdentificacaoMateria']['CodigoMateria'])
        except KeyError:
            CodigoMateria = None
            
        try:
            DescricaoIdentificacaoMateria = item['IdentificacaoMateria']['DescricaoIdentificacaoMateria']
        except KeyError:
            DescricaoIdentificacaoMateria = None
        
        try:
            DescricaoSubtipoMateria = item['IdentificacaoMateria']['DescricaoSubtipoMateria']
        except KeyError:
            DescricaoSubtipoMateria = None
            
        try:
            IndicadorTramitando = item['IdentificacaoMateria']['IndicadorTramitando']
        except KeyError:
            IndicadorTramitando = None
        
        try:
            NomeCasaIdentificacaoMateria = item['IdentificacaoMateria']['NomeCasaIdentificacaoMateria']
        except KeyError:
            NomeCasaIdentificacaoMateria = None
            
        try:
            NomeCasaIdentificacaoMateria = item['IdentificacaoMateria']['NomeCasaIdentificacaoMateria']
        except KeyError:
            NomeCasaIdentificacaoMateria = None
            
        try:
            NumeroMateria = str(item['IdentificacaoMateria']['NumeroMateria'])
        except KeyError:
            NumeroMateria = None
        
        try:
            DataApresentacao = item['DadosBasicosMateria']['DataApresentacao']
        except KeyError:
            DataApresentacao = None
        
        try:
            DataLeitura = item['DadosBasicosMateria']['DataLeitura']
        except KeyError:
            DataLeitura = None
            
        try:
            EmentaMateria = item['DadosBasicosMateria']['EmentaMateria'].strip()
        except KeyError:
            EmentaMateria = None
        
        try:
            ExplicacaoEmentaMateria = item['DadosBasicosMateria']['ExplicacaoEmentaMateria']
        except KeyError:
            ExplicacaoEmentaMateria = None
        
        try:
            IndicadorComplementar = item['DadosBasicosMateria']['IndicadorComplementar']
        except KeyError:
            IndicadorComplementar = None
        
        try:
            IndicadorComplementar = item['DadosBasicosMateria']['IndicadorComplementar']
        except KeyError:
            IndicadorComplementar = None
        
        try:
            DescricaoNatureza = item['DadosBasicosMateria']['NaturezaMateria']['DescricaoNatureza']
        except KeyError:
            DescricaoNatureza = None
            
        try:
            NomeAutor = item['AutoresPrincipais']['AutorPrincipal']['NomeAutor']
        except KeyError:
            NomeAutor = None
        except TypeError:
            NomeAutor = None
            
        try:
            IndicadorOutrosAutores = item['AutoresPrincipais']['AutorPrincipal']['IndicadorOutrosAutores']
        except KeyError:
            IndicadorOutrosAutores = None
        except TypeError:
            IndicadorOutrosAutores = None
            
        try:
            CodigoParlamentar = str(item['AutoresPrincipais']['AutorPrincipal']['IdentificacaoParlamentar']['CodigoParlamentar'])
        except KeyError:
            CodigoParlamentar = None
        except TypeError:
            CodigoParlamentar = None
        
        try:
            NomeParlamentar = item['AutoresPrincipais']['AutorPrincipal']['IdentificacaoParlamentar']['NomeParlamentar'] 
        except KeyError:
            NomeParlamentar = None
        except TypeError:
            NomeParlamentar = None
              
        try:
            NomeCompletoParlamentar = item['AutoresPrincipais']['AutorPrincipal']['IdentificacaoParlamentar']['NomeCompletoParlamentar']
        except KeyError:
            NomeCompletoParlamentar = None
        except TypeError:
            NomeCompletoParlamentar = None
        
        try:
            UfParlamentar = item['AutoresPrincipais']['AutorPrincipal']['IdentificacaoParlamentar']['UfParlamentar']
        except KeyError:
            UfParlamentar = None
        except TypeError:
            UfParlamentar = None
            
        try:
            DataSituacao = str(item['SituacaoAtual']['Autuacoes']['Autuacao']['Situacao']['DataSituacao'])
        except KeyError:
            DataSituacao = None
        except TypeError:
            DataSituacao = None
        
        try:
            DescricaoSituacao = item['SituacaoAtual']['Autuacoes']['Autuacao']['Situacao']['DescricaoSituacao']
        except KeyError:
            DescricaoSituacao = None
        except TypeError:
            DescricaoSituacao = None
        
        dicionario = {"AnoMateria": AnoMateria, 
                      "CodigoMateria": CodigoMateria, 
                      "DescricaoIdentificacaoMateria": DescricaoIdentificacaoMateria, 
                      "DescricaoSubtipoMateria": DescricaoSubtipoMateria, 
                      "IndicadorTramitando": IndicadorTramitando, 
                      "NomeCasaIdentificacaoMateria": NomeCasaIdentificacaoMateria, 
                      "NumeroMateria": NumeroMateria,
                      "ApelidoMateria": ApelidoMateria, 
                      "DataApresentacao": DataApresentacao, 
                      "DataLeitura": DataLeitura,
                      "EmentaMateria": EmentaMateria,
                      "ExplicacaoEmentaMateria": ExplicacaoEmentaMateria,
                      "IndicadorComplementar": IndicadorComplementar,
                      "DescricaoNatureza": DescricaoNatureza,
                      "NomeAutor": NomeAutor,
                      "IndicadorOutrosAutores": IndicadorOutrosAutores,
                      "CodigoParlamentar": CodigoParlamentar,
                      "NomeParlamentar": NomeParlamentar,
                      "NomeCompletoParlamentar": NomeCompletoParlamentar,
                      "UfParlamentar": UfParlamentar,
                      "DescricaoSituacao": DescricaoSituacao
                      }
        projetos.append(dicionario)
```

```python
# Cria dataframe a partir dos dados coletados
```

```python
df_projetos_api = pd.DataFrame(projetos)
```

```python
df_projetos_api = df_projetos_api.drop_duplicates(['CodigoMateria'], keep='last')
```

Vasculha a API para coletar os outros nomes

```python
# É o item MateriaService 
```

```python
prefixo = 'https://legis.senado.leg.br/dadosabertos/materia/'
```

```python
projetos = []

for num, row in mais_autores.iterrows():
    codigo = str(row['CodigoMateria'].strip())
    DescricaoIdentificacaoMateria = row['DescricaoIdentificacaoMateria'].strip()

        
    url = prefixo + codigo
    print (url)
    
    try:
        r = requests.get(url, headers=headers)
    except requests.exceptions.HTTPError as errh:
        print ("Http Error:",errh)
    except requests.exceptions.ConnectionError as errc:
        print ("Error Connecting:",errc) 
    except requests.exceptions.Timeout as errt:
        print ("Timeout Error:",errt)
    except requests.exceptions.RequestException as err:
        print ("OOps: Something Else",err)

    projects = r.json()
    
    
    for item in projects["DetalheMateria"]["Materia"]["Autoria"]["Autor"]:
        #print(item)               
                  
        try:
            CodigoParlamentar = item['IdentificacaoParlamentar']['CodigoParlamentar']
        except KeyError:
            CodigoParlamentar = None
        except TypeError:
            CodigoParlamentar = None
            
        try:
            NomeCompletoParlamentar = item['IdentificacaoParlamentar']['NomeCompletoParlamentar']
        except KeyError:
            NomeCompletoParlamentar = None
        except TypeError:
            NomeCompletoParlamentar = None
            
        #print(CodigoParlamentar)
        
        dicionario = {"CodigoParlamentar": CodigoParlamentar, 
                      "NomeCompletoParlamentar": NomeCompletoParlamentar, 
                      "CodigoMateria": codigo,
                      "DescricaoIdentificacaoMateria": DescricaoIdentificacaoMateria
                      }
        projetos.append(dicionario)
```

```python
df_mais_autores = pd.DataFrame(projetos)
df_mais_autores.info()
```
